```python
import json
import copy
import datetime
import re
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferMemory
from langchain.schema import SystemMessage
from langchain_google_genai import ChatGoogleGenerativeAI

# ---------------------------
# Helper Functions
# ---------------------------

def classify_instruction(user_text: str) -> str:
    """
    Ask the LLM to classify the user's instruction into one of three categories:
    'confirm', 'modify', or 'keep'.
    """
    prompt = (
        "Classify the following instruction regarding a schedule update into one of three categories: "
        "confirm, modify, or keep. Respond with only one word: confirm, modify, or keep.\n"
        f"Instruction: '{user_text}'"
    )
    response = llm.predict(prompt)
    response_lower = response.lower()
    if "confirm" in response_lower:
        return "confirm"
    elif "modify" in response_lower:
        return "modify"
    elif "keep" in response_lower:
        return "keep"
    return "unknown"

def extract_time(text: str) -> str:
    """
    Extract a time string from text in the form HH:MM or HH:MM:SS.
    """
    match = re.search(r'(\d{1,2}:\d{2}(?::\d{2})?)', text)
    if match:
        return match.group(1)
    return None

def extract_checkpoint_and_time(text: str):
    """
    Attempt to extract the intended checkpoint and a time value from the user's text.
    If the phrase "last checkpoint" is found, assume checkpoint "D" (for our example).
    Otherwise, try to find a phrase like "checkpoint X".
    """
    checkpoint = None
    lower_text = text.lower()
    if "last checkpoint" in lower_text:
        checkpoint = "D"  # In our example, D is the last checkpoint.
    else:
        m = re.search(r'checkpoint\s+([A-Z])', text, re.IGNORECASE)
        if m:
            checkpoint = m.group(1).upper()
    time_str = extract_time(text)
    return checkpoint, time_str

# ---------------------------
# Simulation Data
# ---------------------------

planned_schedule_json = """
{
  "truck_schedule": {
    "truck_id": "Truck_001",
    "route": ["A", "B", "C", "D"],
    "planned_checkpoints": [
      {
        "checkpoint": "A",
        "expected_time": "2025-02-07T08:00:00",
        "location": {"latitude": 40.7128, "longitude": -74.0060}
      },
      {
        "checkpoint": "B",
        "expected_time": "2025-02-07T09:00:00",
        "location": {"latitude": 40.7306, "longitude": -73.9352}
      },
      {
        "checkpoint": "C",
        "expected_time": "2025-02-07T10:00:00",
        "location": {"latitude": 40.7580, "longitude": -73.9855}
      },
      {
        "checkpoint": "D",
        "expected_time": "2025-02-07T11:00:00",
        "location": {"latitude": 40.7484, "longitude": -73.9857}
      }
    ]
  }
}
"""

real_time_status_json = """
{
  "real_time_status": {
    "truck_id": "Truck_001",
    "current_checkpoint": "B",
    "current_time": "2025-02-07T09:30:00",
    "current_location": {"latitude": 40.7306, "longitude": -73.9352}
  }
}
"""

# ---------------------------
# System Message for Context
# ---------------------------

system_message = f"""
You are Agent 1, an AI responsible for monitoring an integrated logistics schedule in real time.
Below is the planned schedule and the real-time data received from a truck's IoT devices.

Planned Schedule:
{planned_schedule_json}

Real-Time Data:
{real_time_status_json}

Your task is to immediately analyze the schedule and the real-time data. Identify any discrepancies (for example, delays at checkpoints) and inform the user. 
If a delay is detected, automatically propose an updated schedule by adding the delay to the current and subsequent checkpoints, and show it to the user.
The user may then choose to confirm (in any words that indicate confirmation), modify it (by changing one or more checkpoint times), or keep the original schedule.
Also, if the user types "finalize", ask for manual update details with confirmation before applying the change and simulating an email notification.
Please provide detailed and actionable responses.
"""

# ---------------------------
# Setting up the Gemini LLM via LangChain
# ---------------------------

llm = ChatGoogleGenerativeAI(
    model="gemini-pro",      # Specify your Gemini model
    api_key="AIzaSyApmKwDRSf_A5Pp1ZjM1NXUN1aRfCJUXos",  # Replace with your actual API key
    temperature=0.5
)

# ---------------------------
# Create a ConversationChain with Memory
# ---------------------------

memory = ConversationBufferMemory(memory_key="history", return_messages=True)
memory.chat_memory.messages.append(SystemMessage(content=system_message))
conversation = ConversationChain(llm=llm, memory=memory, verbose=True)

# ---------------------------
# Automatic Initial Analysis
# ---------------------------

initial_query = "Please analyze the planned schedule and the real-time data, and let me know if there is any discrepancy."
initial_response = conversation.predict(input=initial_query)
print("Agent 1:", initial_response)

# ---------------------------
# Delay Detection and Proposal Generation
# ---------------------------

pending_proposal = None       # Basic info about the proposal
pending_final_schedule = None # Holds the full proposed updated schedule (a dictionary)

try:
    planned = json.loads(planned_schedule_json)
    real_time = json.loads(real_time_status_json)
except Exception as e:
    print("Error loading JSON data:", e)
    planned = {}
    real_time = {}

current_checkpoint = real_time.get("real_time_status", {}).get("current_checkpoint")
real_time_current_time_str = real_time.get("real_time_status", {}).get("current_time")
if current_checkpoint and real_time_current_time_str:
    real_time_current_time = datetime.datetime.fromisoformat(real_time_current_time_str)
    # Locate the checkpoint index in the planned schedule.
    index = None
    for i, cp in enumerate(planned.get("truck_schedule", {}).get("planned_checkpoints", [])):
        if cp.get("checkpoint") == current_checkpoint:
            index = i
            break
    if index is not None:
        planned_time_str = planned["truck_schedule"]["planned_checkpoints"][index]["expected_time"]
        planned_time = datetime.datetime.fromisoformat(planned_time_str)
        if real_time_current_time > planned_time:
            delay = real_time_current_time - planned_time
            # Create an updated schedule by adding the delay to the delayed checkpoint and all subsequent ones.
            updated_schedule = copy.deepcopy(planned)
            cps = updated_schedule["truck_schedule"]["planned_checkpoints"]
            for j in range(index, len(cps)):
                orig_time = datetime.datetime.fromisoformat(cps[j]["expected_time"])
                new_time = orig_time + delay
                cps[j]["expected_time"] = new_time.isoformat()
            pending_proposal = {"checkpoint": current_checkpoint, "delay": str(delay)}
            pending_final_schedule = updated_schedule
            proposal_message = (
                f"I detected a delay of {delay} at checkpoint {current_checkpoint}.\n"
                "Based on this, I propose the following updated schedule (the delay is added to the delayed checkpoint and all subsequent checkpoints):\n\n"
                f"{json.dumps(pending_final_schedule, indent=2)}\n\n"
                "Actions:\n"
                "- If you want to confirm the updated schedule, please indicate confirmation in your own words.\n"
                "- If you want to modify the updated schedule (for example, change the expected time for a checkpoint), please express that.\n"
                "- If you want to keep the original schedule, please indicate that as well.\n"
                "You can also type 'finalize' to apply manual changes and send an email notification."
            )
            print("\nAgent 1 Proposal:")
            print(proposal_message)
    else:
        print("Current checkpoint not found in the planned schedule.")
else:
    print("Insufficient real-time data to detect delays.")

# ---------------------------
# Interactive Conversation Loop
# ---------------------------

print("\nStarting interactive conversation with Agent 1.")
print("Type your questions or suggestions about the schedule.")
print("Type 'finalize' to manually apply an update.")
print("Type 'exit' to quit.\n")

while True:
    user_input = input("User: ").strip()
    lower_input = user_input.lower()

    # --- Pending Proposal Branch: Use LLM classification of user instruction ---
    if pending_final_schedule is not None:
        action = classify_instruction(user_input)
        if action == "confirm":
            # Before finalizing, show the default email content and allow user to modify it.
            default_email = f"""
Hello Team,

Please be advised that the schedule for Truck_001 has been updated.
The updated schedule is as follows:

{json.dumps(pending_final_schedule, indent=2)}

Please adjust your operations accordingly.

Best regards,
Agent 1
"""
            print("\nDefault Email Notification to be sent:")
            print(default_email)
            email_modification = input("If you want to modify the email content, type your modifications now (or press Enter to keep it unchanged): ")
            if email_modification.strip():
                default_email = email_modification.strip()
            # Simulate sending the email.
            email_recipients = ["production_manager@example.com", "logistics_team@example.com"]
            email_subject = "Updated Logistics Schedule for Truck_001"
            print("\nSimulated Email Notification:")
            print("Recipients:", ", ".join(email_recipients))
            print("Subject:", email_subject)
            print("Body:", default_email)
            print("\nFinal Updated Schedule:")
            print(json.dumps(pending_final_schedule, indent=2))
            # Save the updated schedule globally.
            planned_schedule_json = json.dumps(pending_final_schedule)
            pending_proposal = None
            pending_final_schedule = None
            continue

        elif action == "modify":
            # Attempt to extract checkpoint and time from the user's message.
            cp_extracted, new_time_extracted = extract_checkpoint_and_time(user_input)
            if cp_extracted is None or new_time_extracted is None:
                # Fallback: ask the user explicitly.
                cp_extracted = input("Enter the checkpoint you want to modify (e.g., C): ").strip()
                new_time_extracted = input("Enter the new expected time for that checkpoint (HH:MM or HH:MM:SS): ").strip()
            else:
                print(f"Extracted modification: checkpoint {cp_extracted}, new time {new_time_extracted}")
            # Update the pending_final_schedule with the extracted data.
            found = False
            cps = pending_final_schedule["truck_schedule"]["planned_checkpoints"]
            for cp_item in cps:
                if cp_item.get("checkpoint") == cp_extracted:
                    old_time = cp_item["expected_time"]
                    date_part = old_time.split("T")[0]
                    # If new_time_extracted has no seconds, append :00.
                    if len(new_time_extracted.split(":")) == 2:
                        new_time_extracted_full = new_time_extracted + ":00"
                    else:
                        new_time_extracted_full = new_time_extracted
                    new_full_time = f"{date_part}T{new_time_extracted_full}"
                    cp_item["expected_time"] = new_full_time
                    found = True
                    print(f"\nModified checkpoint {cp_extracted}: expected time changed from {old_time} to {new_full_time}.")
                    break
            if not found:
                print(f"Checkpoint {cp_extracted} was not found in the proposed schedule.")
            print("\nModified Proposed Schedule:")
            print(json.dumps(pending_final_schedule, indent=2))
            print("If this modified schedule is acceptable, please indicate confirmation (in your own words).")
            continue

        elif action == "keep":
            print("Retaining the original schedule; no changes will be applied.")
            pending_proposal = None
            pending_final_schedule = None
            continue

        else:
            # If the classification is unknown, simply pass the message to the conversation chain.
            response = conversation.predict(input=user_input)
            print("Agent 1:", response)
            continue

    # --- Manual Finalization (if no pending proposal) ---
    if lower_input == "finalize":
        checkpoint_to_update = input("Enter the checkpoint to update (e.g., D): ").strip()
        new_time = input("Enter the new expected time (ISO format, e.g., 2025-02-07T11:30:00): ").strip()
        confirm_manual = input(f"Confirm update of checkpoint {checkpoint_to_update} to new time {new_time}? (yes/no): ").strip().lower()
        if confirm_manual != "yes":
            print("Update canceled. Returning to conversation.")
            continue

        updated_schedule = json.loads(planned_schedule_json)
        found = False
        for checkpoint in updated_schedule["truck_schedule"]["planned_checkpoints"]:
            if checkpoint["checkpoint"] == checkpoint_to_update:
                old_time = checkpoint["expected_time"]
                checkpoint["expected_time"] = new_time
                found = True
                print(f"\nUpdated checkpoint {checkpoint_to_update}: expected time changed from {old_time} to {new_time}.")
                break
        if not found:
            print(f"Checkpoint {checkpoint_to_update} was not found in the schedule.")
            continue

        default_email = f"""
Hello Team,

Please be advised that the schedule for Truck_001 has been updated.
Checkpoint {checkpoint_to_update} expected time is now {new_time}.

Please adjust your operations accordingly.

Best regards,
Agent 1
"""
        print("\nDefault Email Notification to be sent:")
        print(default_email)
        email_modification = input("If you want to modify the email content, type your modifications now (or press Enter to keep it unchanged): ")
        if email_modification.strip():
            default_email = email_modification.strip()

        email_recipients = ["production_manager@example.com", "logistics_team@example.com"]
        email_subject = "Updated Logistics Schedule for Truck_001"
        print("\nSimulated Email Notification:")
        print("Recipients:", ", ".join(email_recipients))
        print("Subject:", email_subject)
        print("Body:", default_email)
        print("\nFinal Updated Schedule:")
        print(json.dumps(updated_schedule, indent=2))
        planned_schedule_json = json.dumps(updated_schedule)
        continue

    elif lower_input == "exit":
        print("Exiting conversation.")
        break

    else:
        # Otherwise, pass the user input to the conversation chain.
        response = conversation.predict(input=user_input)
        print("Agent 1:", response)
```
